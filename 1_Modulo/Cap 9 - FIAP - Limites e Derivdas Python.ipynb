```python
#Bibliotecas

import numpy as np #calculo numerico
import sympy as smp #calculos limites e derivadas
import matplotlib.pyplot as plt #grafico
```

## Limites

```python
 x = smp.Symbol('x') #informar que x é um simbulo
```

```python
funcao = ((x**2 + x - 2) / (x - 1))
```

```python
funcao
```

```python
smp.limit(funcao, x, 1)
```

```python
smp.plot(funcao)
```

 - Estudo de limites da função logística

```python
x = smp.Symbol('x')
x
```

```python
funcao = 1/ (1+smp.exp(-x))
funcao
```

```python
# limite da funcao quando x tende a zero

smp.limit(funcao, x, 0)
```

```python
# limit da funcao quando x tende ao infinito

smp.limit(funcao, x, smp.oo)
```

```python
# limit da funcao quando x tende ao infinito

smp.limit(funcao, x, -smp.oo)
```

###Derivadas

```python
x = smp.Symbol('x')
```

```python
f = x**2
f
```

```python
smp.diff(f) #derivada de x**2
```

- Gráficos

```python
funcao_original = x**2
funcao_derivada = smp.diff(funcao_original)

funcao_original, funcao_derivada
```

```python
# nao funciona com matematica simbolica

funcao_original(2)
```

```python
funcao_original = smp.lambdify(x, funcao_original)
funcao_derivada = smp.lambdify(x, funcao_derivada)
```

```python
# funciona com matematica computacional
funcao_original(2)
```

```python
# grafica da funcao original
y = [funcao_original(i) for i in range(-100, 100)]
plt.plot(y)
```

```python
# grafica da funcao derivada #normal ser uma funcao linear


y_d = [funcao_derivada(i) for i in range(-100, 100)]
plt.plot(y_d)
```
