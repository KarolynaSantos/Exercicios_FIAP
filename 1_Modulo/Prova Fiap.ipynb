```python
# modulos
import pandas as pd
```

```python
df = pd.read_csv('vendas_linha_petshop_2019.csv',
                 sep = ';',
                 decimal = ',',
                 encoding = 'latin1')

df.head()
```

```python
# Contar os valores ausentes em cada coluna
quantidade_missing = df['quantidade'].isna().sum()
valor_total_bruto_missing = df['valor_total_bruto'].isna().sum()

# Calcular o total de entradas
total_entries = len(df)

# Calcular o percentual de dados missing
percent_quantidade_missing = (quantidade_missing / total_entries)
percent_valor_total_bruto_missing = (valor_total_bruto_missing / total_entries)

print(percent_quantidade_missing)
print(percent_valor_total_bruto_missing)
```

```python
dados_norte = df[df['regiao_pais'] == 'Norte']

# Calcular a média dos valores totais brutos
media_valor_total_bruto_norte = dados_norte['valor_total_bruto'].mean()

print("Média dos valores totais brutos para a região Norte: {:.2f}".format(media_valor_total_bruto_norte))
```

```python
# Filtrar os dados para a região Norte
dados_norte = df[df['regiao_pais'] == 'Norte']

# Calcular a moda do índice de desenvolvimento humano (idh) para a região Norte
moda_idh_norte = dados_norte['valor_total_bruto'].mode()[0]

print(moda_idh_norte)
```
